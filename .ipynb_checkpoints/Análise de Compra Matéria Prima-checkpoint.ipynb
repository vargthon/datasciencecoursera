# Importando bibliotecas necess√°rias

```python
import pandas as pd
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import pickle
#%matplotlib inline  
```

# Lendo dados importados

```python
total_fornecedor_2016 = pd.read_csv('compra_fornecedor_total_2016.csv', sep=";")
total_fornecedor_2016['percentagem_lixo'] = (total_fornecedor_2016['totaldesconto'] / total_fornecedor_2016['totalcomprado']) * 100
```

# Mostrando por total de compra

```python
total_fornecedor_2016
```

```python
plt.bar(list(range(0,10)), total_fornecedor_2016[0:10].sort_values(by=['totalcomprado'], ascending=[0])['totalcomprado'])
```

# Total de desconto / Lixo

```python
total_fornecedor_2016.sort_values(by=['percentagem_lixo'], ascending=[1])[0:20]
```

```python
compra_materiaprima = pd.read_csv('compra_materiaprima.csv', sep=";")
compra_materiaprima['desconto_medio'] = (compra_materiaprima['totaldesconto'] / compra_materiaprima['totalcomprado']) * 100
```

```python
compra_materiaprima['mes'] =  compra_materiaprima['dtcompra'].str[5:7]
compra_materiaprima['ano'] =  compra_materiaprima['dtcompra'].str[0:4]
compra_materiaprima['dia'] =  compra_materiaprima['dtcompra'].str[8:10]
(compra_materiaprima[compra_materiaprima['nome'].str.contains('SUCATA - THIA')]
 .sort_values(by=['totalcomprado','ano', 'desconto_medio'], ascending=[0,1,1])
 .groupby(['nome', 'mes', 'desconto_medio']).sum())
```

```python
noclaucio_compra = compra_materiaprima[compra_materiaprima["nome"].str.contains('NOCLA')]
```

```python
print("Valor Total: %.2f" % noclaucio_compra["totalcomprado"].sum())
```
